```python
import pandas as pd
from random import randint
```

```python
df = pd.DataFrame("Day 17, Capstone Project Vocab Builder/Flasher/data/french_words.csv")
#df = pd.read_csv("data/french_words.csv")
```

```python
def randomWordGen():
    
    indexRand = randint(0, len(df))
    
    
    return {"ind": indexRand, "fr": df.iloc[indexRand, 0], "eng": df.iloc[indexRand, 1]}
```

```python
def checker(ans, boolAns):
    
    global df
    
    if(boolAns):
        df = df.drop(ans["ind"]).reset_index(drop = True)
        
    else:
        pass
```

```python
checker(randomWordGen(), False)
df
```
