```python
import pandas as pd
from googletrans import Translator, constants
from pprint import pprint
```

```python
with open("de_full.txt", encoding = "Latin1") as file:
    
    data = file.read()
    
```

```python
data = data.split("\n")
```

```python
corpus = []

for i in data:
    corpus.append(i.split(" "))
    
corpus
```

```python
df = pd.DataFrame(corpus)
```

```python
df = df.drop(len(df) - 1)
```

```python
df.to_csv("final Corpus German.csv")
```
